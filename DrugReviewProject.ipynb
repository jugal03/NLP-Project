### Predicting results

```python
from nltk.corpus import stopwords
def transform(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    y = []
    r = []
    for i in text:
        if i.isalnum():
            y.append(i)
    text = y[:]
    y.clear()
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    text = y[:]
    y.clear()
    for i in text:
        y.append(ps.stem(i))
    for i in y:
        r.append(lm.lemmatize(i))
    return " ".join(r)
```

```python
 def recommend(drug,review):
    text = drug + ' ' + review
    text = transform(text)
    l = [text]
    text = count_vectorizer.transform(l)
    pred = model_r.predict(text)[0]
    return pred
```
